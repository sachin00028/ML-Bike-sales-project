Data Preprocessing

```python
np.round(df['ex_showroom_price'],2)
```

```python
dummy=pd.get_dummies(df[['owner','seller_type',]],drop_first=True)
dummy
```

```python
dfe=pd.concat([df,dummy],axis=1)
dfe
```

```python
dfe=dfe.drop(['seller_type','owner'],axis=1)
dfe
```

```python
# Since Hero honda is the brand which has been changed. So it is renamed to a new model name as Hero

def bike_model(model_name,del_hero_honda=False):
  lst_model=[]
  if del_hero_honda:
    for i in dfe['name']:
      if model_name in i and 'Hero' not in i:
        lst_model.append(i)
    return lst_model

  else:
    for i in dfe['name']:
      if model_name in i:
        lst_model.append(i)
    return lst_model
```

```python
# Categorising each bike name as popular brands

royal_enfield = bike_model('Royal Enfield')
ktm = bike_model('KTM')
yamaha = bike_model('Yamaha')
honda = bike_model('Honda',del_hero_honda=True)
suzuki = bike_model('Suzuki')
hero = bike_model('Hero')
tvs = bike_model('TVS')
bajaj = bike_model('Bajaj')
```

```python
def brand(i):
  if i in royal_enfield:
    return 'Royal Enfield'
  elif i in ktm:
    return 'KTM'
  elif i in yamaha:
    return 'Yamaha'
  elif i in honda:
    return 'Honda'
  elif i in suzuki:
    return 'Suzuki'
  elif i in hero:
    return 'Hero'
  elif i in tvs:
    return 'TVS'
  elif i in bajaj:
    return 'Bajaj'
  else:
    return 'Others'
```

```python
# Categorising each popular brands under one head as Brands

dfe['brand']=dfe['name'].apply(lambda x:brand(x))
```

```python
dfe.head()
```

```python
dfe=dfe.drop(['name'],axis=1)
dfe.head()
```

```python
dummy1=pd.get_dummies(dfe[['brand']],drop_first=True)
dummy1.head()
```

```python
df1=pd.concat([dfe,dummy1],axis=1)
df1.head()
```

```python
df1=df1.drop(['brand'],axis=1)
df1
# df1 is now a pure dataset which had been finished with all the preprocessing steps
```

```python
x=df1.iloc[:,1:].values
y=df1.iloc[:,0].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_absolute_percentage_error,r2_score
print('MAE % ',mean_absolute_percentage_error(y_test,y_pred))
print('r2 score ',r2_score(y_test,y_pred))
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=42)

regressor = RandomForestRegressor(n_estimators=100,max_depth=10,random_state=42)
regressor.fit(x_train,y_train)

y_pred =regressor.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_percentage_error,r2_score
print('MAE % ',mean_absolute_percentage_error(y_test,y_pred))
print('r2 score ',r2_score(y_test,y_pred))
```
